# Data Preparation

## Load the Data

```
raw_data = pd.read_csv("data.csv", header=None)
raw_data.head()
```

```
raw_data = raw_data.iloc[1:].copy()
raw_data.head()
```

```
raw_data['y'] = raw_data['y'].astype('int')
for i in raw_data['y']:
    print(type(i))
    break
```

## Split the Dataset into Training Set & Test Set

```
# Percentage to split by for training
perc = 90

# Set for training the model
data = raw_data.head(int(len(raw_data)*(perc/100))).copy()

# Set for testing the model later to get the real accuracy
test_set = raw_data.tail(int(len(raw_data)*((100-perc)/100))).copy()
```

```
data
```

## Format Data for Training

```
#Take all column values except for the "y" column and the useless "Unnamed: 0" Column as our Continuous x Variable
cont_names_data = list(data.columns.values)
cont_names_data.remove(0)
cont_names_data.remove('y')

print(cont_names_data)
```

```
"""
Old Code - Our model does not use Categorical Variables, only Continuous Variables
cat_names_data = []
for i in data["Unnamed: 0"]:
    cat_names_data.append(i)
print(cat_names_data)

"""
```

```
# Only need to detect Seizure or no Seizure - 1 means seizure is there, 0 means that there is no seizure
data['y'].replace(2, 0, inplace=True)
data['y'].replace(3, 0, inplace=True)
data['y'].replace(4, 0, inplace=True)
data['y'].replace(5, 0, inplace=True)
```

```
# Print new data
data
```

# Making and Training the Tabular Model

##### Tutorial: https://docs.fast.ai/tabular.learner.html
##### Tutorial: https://colab.research.google.com/github/fastai/fastai/blob/master/nbs/44_tutorial.tabular.ipynb

## Make the DataLoaders

```
#Split the data into the training and validation set
splits = RandomSplitter(valid_pct=0.2, seed=None)(range_of(data))

#Use TabularPandas function which will take in all necessary data in order to train model - Using Continuous Variables as well as Dependent Y Variable which has the output
to = TabularPandas(
    data,
    cont_names = cont_names_data,
    procs = [Normalize, FillMissing, Categorify],
    splits = splits,
    y_block = CategoryBlock,
    y_names = 'y'
)

#Take in the Pandas data using a Batch size of 256
dls = to.dataloaders(bs=256)
dls.show_batch()
```

## Training

### Find the best learning rate

```
f1score = F1Score()

# learn = tabular_learner(dls, metrics=[accuracy, error_rate, f1score])
learn = tabular_learner(dls, metrics=[f1score, accuracy, error_rate])
lr_min, lr_steep = learn.lr_find()
print(lr_min)
print(lr_steep)
```

### Train the Model

```
learn.fit(n_epoch=200, lr=lr_min)
```

```
learn.show_results()
```

### Save the Model

```
name = "EEG_torch"
# save_format = int(input("Which format to save the model:\n\t1. .pt - PyTorch\n\t2. pkl - FastAi\n1 or 2:"))
save_format = 1
# Try both of them and let me know which one works
if save_format == 1:
    torch.save(learn, name + ".pt")
    print("Saved the model using torch")
elif save_format == 2:
    learn.export(name + ".pkl")
    print("Saved the model using FastAi")
else:
    print("Did not work")
```
